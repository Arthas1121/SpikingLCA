# Spiking3/Rate4 comparison

A summary notebook for a comparison of the rate4 and spiking3 LCA implementations. If the precomputed results can not be loaded, go to the original notebooks and run them.

## Things I notice

* The spiking implementation is on par with the rate implementation up until 5-8 neurons are active. After this the rate-based LCA slowly converges towards 0 error while the spiking-based LCA can never get below a certain "cieling" error. In fact typically as you would have more neurons active, the error starts increasing.

```python
%load_ext autoreload
%autoreload 2

import neurovivo.plotting.plotting as plt
import numpy as np
import neurovivo.common as cmn
import matplotlib.pyplot as pyl
from LCA.LCA_plotting import plot_rate_spiking_comparison


PATH_FIGURES = "./figures/"
PATH_RESULTS_RATE = cmn.HOME+"/data/LCA/rate4"
PATH_RESULTS_SPIKING = cmn.HOME+"/data/LCA/spiking3"
cmn.mkdir_p(PATH_FIGURES)
```

### Loading the precomputed results

```python
rk4 = cmn.load_pickle(PATH_RESULTS_RATE + "/k=4.pkl")
rk10 = cmn.load_pickle(PATH_RESULTS_RATE + "/k=10.pkl")
rk16 = cmn.load_pickle(PATH_RESULTS_RATE + "/Ndic=64_k=16.pkl")
sk4 = cmn.load_pickle(PATH_RESULTS_SPIKING + "/k=4.pkl")
sk10 = cmn.load_pickle(PATH_RESULTS_SPIKING + "/k=10.pkl")
sk16 = cmn.load_pickle(PATH_RESULTS_SPIKING + "/Ndic=64_k=16.pkl")
```

```python
pyl.figure(figsize=(18,4))
pyl.subplot(1,3,1)
plot_rate_spiking_comparison(rk4, sk4)
pyl.title(r"$N_{dic}$=18, $N_{dic}$=12, k=4")
pyl.xlim(0.6, 1.)
pyl.subplot(1,3,2)
plot_rate_spiking_comparison(rk10, sk10)
pyl.title(r"$N_{dic}$=18, $N_{dic}$=12, k=10")
pyl.xlim(0.4, 1.)
pyl.ylabel("")
pyl.subplot(1,3,3)
plot_rate_spiking_comparison(rk16, sk16)
pyl.title(r"$N_{dic}$=64, $N_{dic}$=16, k=16")
pyl.xlim(0.6, 1.)
pyl.ylabel("")
leg = pyl.legend(loc="best")
leg.draw_frame(False)
pyl.savefig(PATH_FIGURES + "spiking3-rate4-comparison.png")
```
