## A spike-based example of the LCA dynamics:

* This time I take the neurons which are spiking.

### Things I notice:

* A bit obvious, but neurons can not give negative coeficients - they are only positive

```python
import neurovivo.plotting.plotting as plt

import numpy as np
import common as cmn
import brian as br

numpy.random.seed(seed=423456789)
N = 10
b = np.random.randn(N,N)
G = (b + b.T)/2
for i in xrange(len(G)):
    G[i,i]=0.
i_stim = np.random.randn(N)

default_parameters={
            "N":10,
            "v_reset":-60.,         #in mV
            "v_thr":-50.,           #in mV
            "tau_m":20,             #in ms
            "v_rest":-60.,          #in mV
            "dt_delta_thr":10,      #in ms
            "v_rev_e":0.,           #in mV
            "v_rev_i":-80.,         #in mV
            "tau_e":5,              #in ms
            "tau_i":5,              #in ms
            "tau_a":100,            #in ms
            "t_refr":5,             #in ms
            "G":G,
            "i_s":i_stim,
            "tau": 20,
            "dt": 0.1,
            "t_end": 1000,
            "threshold": -50,
            "REPORT": True,
            }

def plot_edge(SMs, i):
    figure()
    plot(-60-(1000.*SMs[0][i]), label="Vrest-vm")
    plot(SMs[1][i], label="ge")
    plot(SMs[2][i], label="gi")
    #plot(1000.*SMs[0][i])
    #plot(SMs[3][i], label="a")
    plot(SMs[4][i], label="i_s")
    legend()

def plot_a_evolutions(SMs, N):
    [plot(SMs[3][i][::10], "blue") for i in xrange(N)]
    
PATH_FIGURES = cmn.HOME + "/data/LCA/figures/"
PATH_FIGURES = "./figures/"
cmn.mkdir_p(PATH_FIGURES)
```

```python
def simulate_spiking(parameters=default_parameters):
    x = cmn.Bunch(parameters)
    tau_m_inv = 1./(x.tau*br.ms)
    tau_e_inv = 1./(x.tau_e*br.ms)
    tau_i_inv = 1./(x.tau_i*br.ms)
    tau_a_inv = 1./(x.tau_a*br.ms)

    br.clear(all=True) # clears away all the brian objects from previous simulations.
    
    def thresh_func(values, threshold):
        vals = values.copy()
        vals[(vals<threshold) & (vals>-threshold)]=0.
        return 0.1*vals
    
    eqs = '''
          dvm/dt=((x.v_rest*br.mV-vm) + ge*(x.v_rev_e*br.mV-vm) + gi*(x.v_rev_i*br.mV-vm) + i_s)*tau_m_inv :br.mV
          dge/dt = -ge*tau_e_inv :1
          dgi/dt = -gi*tau_i_inv :1
          da/dt = -a*tau_a_inv   :br.Hz
          i_s                 :br.mV
          '''
        
    simulation_clock = br.Clock(dt=x.dt*br.ms)
    reporting_clock = br.Clock(dt=1000*br.ms)
    recording_clock = br.Clock(dt=.1*br.ms)
    
    P = br.NeuronGroup(x.N, model=eqs, threshold=x.v_thr*br.mV, reset=x.v_reset*br.mV, refractory=x.t_refr*br.ms, clock=simulation_clock)
    P.vm = -60 * br.mV
    P.i_s = 10 * x.i_s * br.mV
    P.a = 0

    G_exc = G.copy()
    G_exc[G_exc<0]=0
    G_inh = G.copy()
    G_inh[G_inh>0]=0
    G_inh = -1*G_inh
    
    Ce = br.Connection(P, P, 'ge')
    Ce.connect(P, P, 0.25*G_exc)
    Ci = br.Connection(P, P, 'gi')
    Ci.connect(P, P, 1.*G_inh)
    
    Ca = br.IdentityConnection(P, P, 'a', weight=tau_a_inv)
            
    if x.REPORT:
        @br.network_operation(clock=reporting_clock, when="start")
        def report(reporting_clock):
            print reporting_clock.t

    M = br.SpikeMonitor(P)
    
    SMs = []
    SM_1 = br.StateMonitor(P, "vm", record=True, timestep=1, clock=recording_clock)
    SM_2 = br.StateMonitor(P, "ge", record=True, timestep=1, clock=recording_clock)
    SM_3 = br.StateMonitor(P, "gi", record=True, timestep=1, clock=recording_clock)
    SM_4 = br.StateMonitor(P, "a", record=True, timestep=1, clock=recording_clock)
    SM_5 = br.StateMonitor(P, "i_s", record=True, timestep=1, clock=recording_clock)
    SMs.append(SM_1)
    SMs.append(SM_2)
    SMs.append(SM_3)
    SMs.append(SM_4)
    SMs.append(SM_5)
    
    if x.REPORT:
        network = br.Network(P, Ce, Ci, Ca, report, M, *SMs)
    else:
        network = br.Network(P, Ce, Ci, Ca, M, *SMs)
        
    # starting the simulation
    network.run(x.t_end*br.ms)

    if x.REPORT:
        print "simulation finished"
    return SMs, M
```

```python
SMs, M = simulate_spiking()
```

```python
br.raster_plot(M)
ylim(0,N)
savefig(PATH_FIGURES + "/spiking_raster_plot.png")
```

```python
I=2
plot(SMs[1][I]*(-1000*SMs[0][I]))
plot(SMs[2][I]*(-80-1000*SMs[0][I]))
plot(SMs[4][I])
plot(-60-1000*SMs[0][I])
plot(1000*SMs[0][I])
```

```python
plot_a_evolutions(SMs, N)
xlabel("time [ms]");
ylabel("a(t)")
savefig(PATH_FIGURES + "/spiking_a_evolution.png")
```
