# Side-question: Can I have an IF neuron which integrates the input?

## What do I find:

The integration in the single neuron definitelly does not come out of the box. The way I do it is I connect the neuron it to itself in the hope it will sustain its own firing rate when presented with no input and intergrate the input otherwise. It turns out the firing rate seems to have two stable points - one at 0 HZ (no spiking) and the other at the high spiking values, close to it's maximal firing rate (defined by the refractory period). See the first figure.

```python
%load_ext ipycache

import matplotlib.pyplot as pyl
import neurovivo.plotting.plotting as plt
import numpy as np
import common as cmn
import brian as br

np.random.seed(seed=123456789)


default_parameters={
            "N":1,
            "v_reset":-60.,         #in mV
            "v_thr":-50.,           #in mV
            "tau_m":5,              #in ms
            "v_rest":-60.,          #in mV
            "dt_delta_thr":10,      #in ms
            "v_rev_e":0.,           #in mV
            "tau_e":20,             #in ms
            "tau_a":50,             #in ms
            "t_refr":1,             #in ms
            "i_s1":0,
            "i_s2":15.,
            "t_stim": 100,
            "stim_start": 100,
            "dt": 0.1,
            "t_end": 1000,
            "w_e":0.1,
            "REPORT": False,
            }

PATH_FIGURES = "./figures/single_neuron_integrator"
cmn.mkdir_p(PATH_FIGURES)


def simulate_spiking(parameters=default_parameters):
    x = cmn.Bunch(parameters)
    tau_m_inv = 1./(x.tau_m*br.ms)
    tau_e_inv = 1./(x.tau_e*br.ms)
    tau_a_inv = 1./(x.tau_a*br.ms)
    
    #br.clear(all=True) # clears away all the brian objects from previous simulations.
        
    eqs = '''
          dvm/dt=((x.v_rest*br.mV-vm) + ge*(x.v_rev_e*br.mV-vm) + i_s)*tau_m_inv :br.mV
          dge/dt = -ge*tau_e_inv :1
          i_s                 :br.mV
          da/dt = -a*tau_a_inv   :br.Hz
          '''

    simulation_clock = br.Clock(dt=x.dt*br.ms)
    reporting_clock = br.Clock(dt=1000*br.ms)
    recording_clock = br.Clock(dt=.1*br.ms)
    
    P = br.NeuronGroup(x.N, model=eqs, threshold=x.v_thr*br.mV, reset=x.v_reset*br.mV, refractory=x.t_refr*br.ms, clock=simulation_clock)
    P.vm = -60 * br.mV

    I = list(x.i_s1*br.mV*np.ones(int(x.stim_start/x.dt)))+list(x.i_s2*br.mV*np.ones(int((x.t_stim)/x.dt)))+list(x.i_s1*br.mV*np.ones(int((x.t_end-x.stim_start-x.t_stim)/x.dt)))
#    P.i_s = br.TimedArray(np.array([x.i_s1*br.mV, x.i_s2*br.mV, x.i_s1*br.mV]))
    P.i_s = br.TimedArray(I, clock=simulation_clock)    
    Ce = br.Connection(P, P, 'ge')
    Ce.connect_one_to_one(P, P, x.w_e)

    Ca = br.IdentityConnection(P, P, 'a', weight=tau_a_inv)
    
    if x.REPORT:
        @br.network_operation(clock=reporting_clock, when="start")
        def report(reporting_clock):
            print reporting_clock.t

    M = br.SpikeMonitor(P)
    
    SMs = []
    SM_1 = br.StateMonitor(P, "vm", record=True, timestep=1, clock=recording_clock)
    SM_2 = br.StateMonitor(P, "ge", record=True, timestep=1, clock=recording_clock)
    SM_3 = br.StateMonitor(P, "i_s", record=True, timestep=1, clock=recording_clock)
    SM_4 = br.StateMonitor(P, "a", record=True, timestep=1, clock=recording_clock)
    SMs.append(SM_1)
    SMs.append(SM_2)
    SMs.append(SM_3)
    SMs.append(SM_4)
    
    if x.REPORT:
        network = br.Network(P, Ce, Ca, report, M, *SMs)
    else:
        network = br.Network(P, Ce, Ca, M, *SMs)
        
    # starting the simulation
    network.run(x.t_end*br.ms)

    if x.REPORT:
        print "simulation finished"
    return SMs, M
```

## Can we even have sustained firing with only self-connected neuron

```python
pars = default_parameters.copy()
pars["t_stim"] = 50
pars["stim_start"] = 150
SMs, M = simulate_spiking(pars)
pars2 = default_parameters.copy()
pars2["t_stim"] = 25
SMs2, M2 = simulate_spiking(pars2)
pars3 = default_parameters.copy()
pars3["stim_start"] = 70
pars3["t_stim"] = 15
SMs3, M3 = simulate_spiking(pars3)
```

```python
time = np.arange(len(SMs[3][0][:]))*0.1
pyl.plot(time, 7000*SMs[2][0][:], c="b")
pyl.plot(time, SMs[3][0][:], c="b")
pyl.plot(time, 7000*SMs2[2][0][:], c="r")
pyl.plot(time, SMs2[3][0][:], c="r")
pyl.plot(time, 7000*SMs3[2][0][:], c="g")
pyl.plot(time, SMs3[3][0][:], c="g")
pyl.xlabel("time [ms]")
pyl.xlim(0,400)
pyl.savefig(PATH_FIGURES + "/integration_result.png")
```

**The self-connected neuron firing rates have two stable points.** Each color represents a run with a different stimulation time window (the boxes). During the stimulation the signals seem to be integrating, however after the signal is turned off, the neuron activity (the wiggley lines) converges to one of the two stable points.
