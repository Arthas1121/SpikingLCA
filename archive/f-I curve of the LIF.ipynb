# Investigation in the f-I curves of the LIF neuron

The f-I curve of the conductance-based LIF neuron should look a bit like the thresholding function T in the LCA model. Here I explore how it depends on the various parameters.

I look at two input types:

* Direct current
* Conductive synaptic input (a regular spiking input of various frequencies)

## Things I find

* Both input types have qualitatively same f-I curve (or at least very a very similar one)
* The slope of the curve after the threshold value is never quite linear. There is some amount of sublinearity. present. And of course there would be the saturation at values determined by the refractory period.
* Changing the membrane threshold value not only changes the min requried input for spiking (the equivalent of the threshold in the LCA model), but it also changes the slope of the curve.

```python
%load_ext ipycache
import neurovivo.plotting.plotting as plt
import matplotlib.pyplot as pyl

import numpy as np
import common as cmn
import brian as br

#numpy.random.seed(seed=123456789)

default_parameters={
            "N":1,
            "v_reset":-60.,         #in mV
            "v_thr":-55.,           #in mV
            "tau_m":20,              #in ms
            "v_rest":-60.,          #in mV
            "v_rev_e":0.,           #in mV
            "tau_e":20,             #in ms
            "tau_a":50,             #in ms
            "t_refr":1,             #in ms
            "i_s":0.,
            "i_noise":0.,
            "stim_freq": 0.,
            "dt": 0.1,
            "t_end": 1000,
            "w_e":0.1,
            }

PATH_FIGURES = "./figures/f-I_curve/"
cmn.mkdir_p(PATH_FIGURES)

def number_of_spikes(M, t_start, t_end):
    """
    Returns list of number of spikes per neuron between t_start and t_end.
    """
    result = []
    for st in xrange(len(M.source)):
        result.append(len(M[st][(M[st]>(t_start/1000.)) & (M[st]<(t_end/1000.))]))
    return np.array(result).T

def simulate_spiking(parameters=default_parameters):
    x = cmn.Bunch(parameters)
    tau_m_inv = 1./(x.tau_m*br.ms)
    tau_e_inv = 1./(x.tau_e*br.ms)
    tau_a_inv = 1./(x.tau_a*br.ms)
    
    br.clear(all=True) # clears away all the brian objects from previous simulations.
        
    eqs = '''
          dvm/dt=((x.v_rest*br.mV-vm) + ge*(x.v_rev_e*br.mV-vm) + i_s + i_noise)*tau_m_inv :br.mV
          dge/dt = -ge*tau_e_inv :1
          i_s                    :br.mV
          i_noise                :br.mV
          da/dt = -a*tau_a_inv   :br.Hz
          '''

    simulation_clock = br.Clock(dt=x.dt*br.ms)
    recording_clock = br.Clock(dt=.1*br.ms)
    
    P = br.NeuronGroup(x.N, model=eqs, threshold=x.v_thr*br.mV, reset=x.v_reset*br.mV, refractory=x.t_refr*br.ms, clock=simulation_clock)
    if x.stim_freq > 0.:
        spiketimes = np.arange(0, x.t_end/1000., 1./x.stim_freq)
        Pext = br.SpikeGeneratorGroup(x.N, spiketimes=(np.zeros(len(spiketimes)),spiketimes), clock=simulation_clock)
        #Pext = br.PoissonGroup(x.N, rates = x.stim_freq*br.Hz, clock=simulation_clock)
        Ce = br.IdentityConnection(Pext, P, 'ge', weight=x.w_e)

    P.vm = -60 * br.mV

    P.i_s = x.i_s * br.mV
    I = x.i_noise*br.mV*np.random.normal(size=int(x.t_end/x.dt))
    P.i_noise = br.TimedArray(I, clock=simulation_clock)

    Ca = br.IdentityConnection(P, P, 'a', weight=tau_a_inv)
    
    M = br.SpikeMonitor(P)
    
    SMs = []
    #SM_1 = br.StateMonitor(P, "vm", record=True, timestep=1, clock=recording_clock)
    #SM_2 = br.StateMonitor(P, "ge", record=True, timestep=1, clock=recording_clock)
    #SM_3 = br.StateMonitor(P, "i_noise", record=True, timestep=1, clock=recording_clock)
    #SM_4 = br.StateMonitor(P, "a", record=True, timestep=1, clock=recording_clock)
    #SMs.append(SM_1)
    #SMs.append(SM_2)
    #SMs.append(SM_3)
    #SMs.append(SM_4)
    
    if x.stim_freq > 0.:
        network = br.Network(P, Pext, Ce, Ca, M, *SMs)
    else:
        network = br.Network(P, Ca, M, *SMs)
        
    # starting the simulation
    network.run(x.t_end*br.ms)

    return SMs, M
```

```python
%%cache /Users/mpelko/data/LCA/tmp/iF_curves_i_ss.pkl thresholds, i_sss, ratess

thresholds = [-59.5, -59, -58, -57, -55, -53, -50]
parms = default_parameters.copy()
ratess = []
i_sss = []
for thr in thresholds:
    print thr
    parms["v_thr"]=thr
    i_ss = [0, parms["v_thr"]-parms["v_rest"]] + list(np.arange(parms["v_thr"]-parms["v_rest"], 20, 0.2))
    i_sss.append(i_ss)
    rates = []
    for i in i_ss:
        parms["i_s"]=i
        SMs, M = simulate_spiking(parms)
        rates.append(number_of_spikes(M, 300, 1000)[0]/0.7)
    ratess.append(rates)
```

```python
pyl.figure(figsize=(6,4))
for i in xrange(len(thresholds)):
    pyl.plot(i_sss[i], ratess[i], label=r"$V_{thr}=$"+" {}".format(thresholds[i]))
leg = pyl.legend(loc="best", prop={"size":12})
leg.draw_frame(False)
pyl.ylabel("Output rate [Hz]")
pyl.xlabel(r"$i_{stim} [mV]$");
pyl.ylim(0,300)
pyl.savefig(PATH_FIGURES + "f-I_on_threshold_curr-stim.png")
```
