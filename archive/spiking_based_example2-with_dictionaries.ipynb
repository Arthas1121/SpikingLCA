## A spike-based example of the LCA dynamics:

* This time I take the neurons which are spiking.

### Things I notice:

* A bit obvious, but neurons can not give negative coeficients - they are only positive.
* I do not have a good way of relating the length of the reconstruction vectors to the target input vectors. For that reason, I resize the reconstruction vectors to the length that yields the smallest error, where error is the euclidian distance from the target input vector. I believe this is also what happens in rate-based LCA.
* There are now many parameters which I am now setting by hand and not in any principled way. I should think about how to get to the more principled way, but I might rather do this in the next stage.
* I get to some overlap between the active neurons in the spiking and the rate-based case if I tune the parameters right (3 out of 5) - not shown here.
* The mean error for taking a random vector and optimize it's length for our dimensionality is about 0.964. NOTE: When we rescale the vector length, we might also invert it's direction (scaling factor can be negative) if this would cauese a better fit.

#### Medium strength of connections

* I can get the error down to somewhere around 0.5 for a reasonable sparsness (8 out of 64). (NOTE: the max. error is 2, mean would be around 0.96)
* I can get the error down to 0.1 or so, but that would be some sort of overfitting already, as we use about 42 active nodes to represent a vector of dimension 16.
* When we maximally lower the threshold, about 80% of the nodes are active. At that point the reconstruction error is rather big (as good as random).

#### Weak strength of connections (0.1 times the medium)

* I can get the error down to somewhere around 0.65 for a reasonable sparsness (8 out of 64).
* I can get the error down to 0.5 or so, but that would be some sort of overfitting already, as we use about 30 active nodes to represent a vector of dimension 16.
* When we maximally lower the threshold, about 40% of all the nodes are active.

#### Strong strength of connections (10 times the medium)

* I can not get a reasonable sparsness at all (I expect due the to big fluctuations of the membrane voltage). I either have about 80% of the nodes active, or 1 node or none. Good example of how overconnectivity can kill your network usability.

```python
%load_ext ipycache

import neurovivo.plotting.plotting as plt

import numpy as np
import common as cmn
import brian as br

numpy.random.seed(seed=123456789)
Ndic = 64
Ndim = 16

# create N normalized dictionaries - nonoptimized
b = np.random.randn(Ndic,Ndim)
norms = np.apply_along_axis(np.linalg.norm, 1, b)
b_n = b / norms.reshape(-1,1)

# create a G matrix from the normalized dictionaries
G = np.zeros((Ndic,Ndic))
for i in xrange(Ndic):
    for j in xrange(i,Ndic):
        G[i][j] = np.dot(b_n[i], b_n[j])
G = G + G.T - 2*np.identity(Ndic)
G[np.abs(G)<0.00000000001] = 0.

# create i_stim based on the random input vector we are sparsifyign

x = np.random.randn(Ndim)
x_norm = 1.*x/np.linalg.norm(x)
i_stim = np.dot(b_n,x_norm)

default_parameters={
            "N":Ndic,
            "v_reset":-60.,         #in mV
            "v_thr":-50.,           #in mV
            "tau_m":20,             #in ms
            "v_rest":-60.,          #in mV
            "dt_delta_thr":10,      #in ms
            "v_rev_e":0.,           #in mV
            "v_rev_i":-80.,         #in mV
            "tau_e":5,              #in ms
            "tau_i":5,              #in ms
            "tau_a":100,            #in ms
            "t_refr":5,             #in ms
            "G":G,
            "i_s":i_stim,
            "tau": 20,
            "dt": 0.1,
            "t_end": 1000,
            "e_factor":0.01,
            "i_factor":0.03,
            "REPORT": False,
            }

PATH_FIGURES = "./figures/"
cmn.mkdir_p(PATH_FIGURES)

def plot_edge(SMs, i):
    figure()
    plot(-60-(1000.*SMs[0][i]), label="Vrest-vm")
    plot(SMs[1][i], label="ge")
    plot(SMs[2][i], label="gi")
    #plot(1000.*SMs[0][i])
    #plot(SMs[3][i], label="a")
    plot(SMs[4][i], label="i_s")
    legend()

def plot_a_evolutions(SMs, N):
    [plot(SMs[3][i][::10], "blue") for i in xrange(N)]

def reconstruction(b_n, M, t_end):
    """
    Returns a reconstructed input vector by using the freq of spiking
    in the last second.
    """
    a = number_of_spikes_last_second(M, t_end)
    rec = np.dot(b_n.T,a)
    return rec

def reconstruction_error(b_n, x, rec, fact=1.):
    assert np.linalg.norm(x)==1.
    if not np.linalg.norm(rec) == 0:
        rec_n = rec/np.linalg.norm(rec)
    else: 
        rec_n = rec
    rec_op = fact*np.dot(rec_n, x) * rec_n
    return np.linalg.norm(x - rec_op)
    
def number_of_active_neurons(M, min_time):
    """
    Returns the number of neurons that spiked since min_time until 
    the end of the simulation.
    """
    n_active = 0
    for st in xrange(len(M.source)):
        if len(M[st][M[st]>min_time/1000.])>0:
            n_active += 1
    return n_active

def number_of_spikes_last_second(M, t_end):
    """
    Returns list of number of spikes per neuron in the last second 
    before t_end.
    """
    result = []
    for st in xrange(len(M.source)):
        result.append(len(M[st][M[st]>(t_end/1000.-1.)]))
    return np.array(result).T
```

```python
def simulate_spiking(parameters=default_parameters):
    x = cmn.Bunch(parameters)
    tau_m_inv = 1./(x.tau*br.ms)
    tau_e_inv = 1./(x.tau_e*br.ms)
    tau_i_inv = 1./(x.tau_i*br.ms)
    tau_a_inv = 1./(x.tau_a*br.ms)

    br.clear(all=True) # clears away all the brian objects from previous simulations.
        
    eqs = '''
          dvm/dt=((x.v_rest*br.mV-vm) + ge*(x.v_rev_e*br.mV-vm) + gi*(x.v_rev_i*br.mV-vm) + i_s)*tau_m_inv :br.mV
          dge/dt = -ge*tau_e_inv :1
          dgi/dt = -gi*tau_i_inv :1
          da/dt = -a*tau_a_inv   :br.Hz
          i_s                 :br.mV
          '''
        
    simulation_clock = br.Clock(dt=x.dt*br.ms)
    reporting_clock = br.Clock(dt=1000*br.ms)
    recording_clock = br.Clock(dt=.1*br.ms)
    
    P = br.NeuronGroup(x.N, model=eqs, threshold=x.v_thr*br.mV, reset=x.v_reset*br.mV, refractory=x.t_refr*br.ms, clock=simulation_clock)
    P.vm = -60 * br.mV
    P.i_s = 1. * x.i_s * br.mV
    P.a = 0

    G_exc = G.copy()
    G_exc[G_exc>0]=0
    G_exc = -1.*G_exc
    G_inh = G.copy()
    G_inh[G_inh<0]=0
    
    Ce = br.Connection(P, P, 'ge')
    Ce.connect(P, P, x.e_factor*G_exc)
    Ci = br.Connection(P, P, 'gi')
    Ci.connect(P, P, x.i_factor*G_inh)
    
    Ca = br.IdentityConnection(P, P, 'a', weight=tau_a_inv)
            
    if x.REPORT:
        @br.network_operation(clock=reporting_clock, when="start")
        def report(reporting_clock):
            print reporting_clock.t

    M = br.SpikeMonitor(P)
    
    SMs = []
    SM_1 = br.StateMonitor(P, "vm", record=True, timestep=1, clock=recording_clock)
    SM_2 = br.StateMonitor(P, "ge", record=True, timestep=1, clock=recording_clock)
    SM_3 = br.StateMonitor(P, "gi", record=True, timestep=1, clock=recording_clock)
    SM_4 = br.StateMonitor(P, "a", record=True, timestep=1, clock=recording_clock)
    SM_5 = br.StateMonitor(P, "i_s", record=True, timestep=1, clock=recording_clock)
    SMs.append(SM_1)
    SMs.append(SM_2)
    SMs.append(SM_3)
    SMs.append(SM_4)
    SMs.append(SM_5)
    
    if x.REPORT:
        network = br.Network(P, Ce, Ci, Ca, report, M, *SMs)
    else:
        network = br.Network(P, Ce, Ci, Ca, M, *SMs)
        
    # starting the simulation
    network.run(x.t_end*br.ms)

    if x.REPORT:
        print "simulation finished"
    return SMs, M
```

## Simple example

```python
pars=default_parameters.copy()
pars["e_factor"] = 10.*pars["e_factor"]
pars["i_factor"] = 10.*pars["i_factor"]
pars["v_thr"] = -59.65
pars["t_end"] = 1000
SMs, M = simulate_spiking(pars)
```

## Strong connectivity sweep

```python
%%cache /Users/mpelko/data/LCA/tmp/spike-based-example2.pkl thresholds, nr_active, errors

pars=default_parameters.copy()
pars["t_end"] = 1000
thresholds = np.linspace(-59.99, -59.7, 20)
nr_active, errors = [], []
for t in thresholds:
    pars["v_thr"] = t
    SMs, M = simulate_spiking(pars)
    rec = reconstruction(b_n, M, 1000)
    errors.append(reconstruction_error(b_n, x_norm, rec, 1.))
    nr_active.append(number_of_active_neurons(M, 0))
```

```python
def plot_reconstruction_error(thresholds, nr_active, error):
    figure(figsize=(12,4))
    subplot(1,2,1)
    plot(thresholds, error, label=r"$\|x-(\phi a)_{op}\|$")
    xlabel(r"threshold ($\lambda$)")
    plot(thresholds, 1.-1.*np.array(nr_active)/Ndic, label=r"sparsness")
    legend(loc="best")
    subplot(1,2,2)
    plot(1.-1.*np.array(nr_active)/Ndic, error, label=r"$\|x-(\phi a)_{op}\|$")
    xlabel("sparsness")
    #plot(sparsnesses, reconstruction_norms, label=r"$\|\phi a\|$")
    legend(loc="best")

plot_reconstruction_error(thresholds, nr_active, errors)
savefig(PATH_FIGURES + "spike-reconstruction-error.png")
```
