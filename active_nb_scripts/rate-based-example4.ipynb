# A rate based equivalent to spiking3 example of the LCA dynamics:

The rate based example, which is the basis for the spiking3. Generative model for surrogate data, positive only output vectors, 

### Things I notice:



```python
%load_ext ipycache
%load_ext autoreload
%autoreload 2

import neurovivo.plotting.plotting as plt
import numpy as np
import neurovivo.common as cmn

from LCA_common import generative_model_data
from LCA_simulate import simulate_rate4 as simulate
from LCA_plotting import plot_reconstruction_error
from LCA_analysis import reconstruction_evaluation

default_parameters={
            "Ndic":64,              # number of dictionaries
            "Ndim":16,              # input dimension (smaller or equal then Ndic)
            "k":4,                  # number of dictionaries used to create the input
            "v_reset":-60.,         # in mV
            "v_thr":-50.,           # in mV
            "tau_m":1,              # in ms
            "v_rest":-60.,          # in mV
            "dt_delta_thr":10,      # in ms
            "v_rev_e":0.,           # in mV
            "v_rev_i":-80.,         # in mV
            "tau_e":5,              # in ms
            "tau_i":10,             # in ms - note theory requires formally tau_i < tau_e
            "t_refr":5,             # in ms
            "dt": 0.1,
            "t_end": 1000,
            "threshold": 0.4,
            "positive_input": True,
            "positive_phi": False,
            "nonnegative":True,
            "thr_type": "hard",     # "soft" or "hard" threshold.
            "input_factor": 1.,
            "thr_coeff": 1.,        
            "seed":None,            # random seed used for the generative model.
            }

PATH_FIGURES = "./figures/rate_based/"
PATH_CACHING = cmn.HOME+"/data/LCA/tmp/"
cmn.mkdir_p(PATH_FIGURES)
PATH_RESULTS = cmn.HOME+"/data/LCA/rate4/"
cmn.mkdir_p(PATH_RESULTS)
```

## input factor sweep

### k = 4

```python
file_name0 = PATH_CACHING + "/rate-based-example4_0.pkl"
```

```python
%%cache $file_name0 a_LCAs0, GMs0, pars0, input_factors0

pars0 = default_parameters.copy()
pars0["seed"] = 12345678
pars0["Ndic"] = 18
pars0["Ndim"] = 12
pars0["k"] = 4
pars0["thr_type"] = "soft"

input_factors0 = [0.1, 0.5, 1., 1.5, 2., 3., 4., 5., 6., 7., 8., 9., 10., 12., 15., 17.5, 20., 25.,]
a_LCAs0, GMs0 = [], []
for inp_f in input_factors0:
    pars0["input_factor"] = inp_f
    a_LCA, GM = simulate(pars0)
    a_LCAs0.append(a_LCA)
    GMs0.append(GM)
```

```python
errors0, nr_active0 = [], []
for a_LCA, GM in zip(a_LCAs0, GMs0):
    analysis = reconstruction_evaluation(a_LCA, GM, pars0)
    errors0.append(analysis["opt_error"])
    nr_active0.append(analysis["nonzero"])

results = {"inp_f": input_factors0,
           "err": errors0,
           "nr_act": nr_active0,
           "pars":pars0,
            }
cmn.save_pickle(PATH_RESULTS+"/k=4.pkl", results)
```

```python
plot_reconstruction_error(input_factors0, nr_active0, errors0, pars0["Ndic"])
savefig(PATH_FIGURES + "rate-reconstruction-error4_0.png")
```
