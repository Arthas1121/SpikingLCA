```python
dict_pos = dict_from_flat_data(weights_pos_data, weights_pos_header)
dict_neg = dict_from_flat_data(weights_neg_data, weights_neg_header)
V1_on_act=np.rollaxis(np.array(V1_on_sp.rates()).reshape(8,8,32),1)
```

```python
def reconstruct_image_from_activities(V1_on, dict_pos, dict_neg):
    image1 = reconstruct_image(V1_on, dict_pos, nx_rel_scale=.25, ny_rel_scale=.25)
    image2 = reconstruct_image(V1_on, dict_neg, nx_rel_scale=.25, ny_rel_scale=.25)
    #return image1, image2, image3, image4, image1+image4-image2-image3
    return image1, image2, image1-image2
```

```python
im1, im2, im3 = reconstruct_image_from_activities(V1_on_act, dict_pos, dict_neg)
```

```python
pyl.figure(figsize=(16,4))
pyl.subplot(1,3,1)
imshow(im1[:,:,0].T, cmap=cm.Greys, interpolation="nearest")
pyl.subplot(1,3,2)
imshow(im2[:,:,0].T, cmap=cm.Greys, interpolation="nearest")
pyl.subplot(1,3,3)
imshow(im3[:,:,0].T, cmap=cm.Greys, interpolation="nearest")
```

```python
rate_act = sparse_to_full(V1_rate_data, V1_rate_header, i)
_, _, im3r = reconstruct_image_from_activities(rate_act, dict_pos, dict_neg)

gan_im = ganglion_image.T
ret_im = np.reshape(np.array(retina_on_sp.rates())-np.array(retina_off_sp.rates()), (32,32));
rec_im = im3[:,:,0].T;
rate_im = im3r[:,:,0].T;

norm_gan_im = (gan_im - np.mean(gan_im))/np.std(gan_im - np.mean(gan_im))
norm_ret_im = (ret_im - np.mean(ret_im))/np.std(ret_im - np.mean(ret_im))
norm_rec_im = (rec_im - np.mean(rec_im))/np.std(rec_im - np.mean(rec_im))
norm_rate_im = (rate_im - np.mean(rate_im))/np.std(rate_im - np.mean(rate_im))
```

```python
pyl.figure(figsize=(16,4))
pyl.subplot(1,5,1)
pyl.title("ganglion_norm")
pyl.imshow(norm_gan_im, cmap=cm.Greys, interpolation="nearest");
pyl.subplot(1,5,2)
pyl.title("retina_norm")
pyl.imshow(norm_ret_im, cmap=cm.Greys, interpolation="nearest");
pyl.subplot(1,5,3)
pyl.title("rate_rec_norm")
pyl.imshow(norm_rate_im, cmap=cm.Greys, interpolation="nearest");
pyl.subplot(1,5,4)
pyl.title("rec_norm")
pyl.imshow(norm_rec_im, cmap=cm.Greys, interpolation="nearest")
pyl.subplot(1,5,5)
pyl.title("diff_gang")
pyl.imshow(norm_gan_im-norm_rec_im, cmap=cm.Greys, interpolation="nearest")
```
