```python
%load_ext ipycache
%load_ext autoreload
%autoreload 2

import neurovivo.plotting.plotting as plt
import matplotlib.pylab as pyl
import numpy as np
import neurovivo.common as cmn
import brian as br

from LCA_simulate import simulate_spiking3 as simulate
from LCA_plotting import plot_a_evolutions, plot_reconstruction_error
from LCA_analysis import get_dynamic_a, get_dynamic_error, number_of_active_neurons, reconstruction_evaluation

default_parameters={
            "Ndic":18,              # number of dictionaries
            "Ndim":12,              # input dimension (smaller or equal then Ndic)
            "k":4,                  # number of dictionaries used to create the input
            "v_reset":-60.,         #in mV
            "v_thr":-55.,           #in mV
            "tau_m":20,              #in ms
            "v_rest":-60.,          #in mV
            "v_rev_e":0.,           #in mV
            "v_rev_i":-80.,         #in mV
            "w_e":0.3/3.,
            "w_i":0.3,            
            "tau_e":5,              #in ms
            "tau_i":5,              #in ms
            "tau_a":100,            #in ms
            "t_refr":.25,             #in ms
            "dt": 0.1,
            "t_end": 1000,
            "input_factor": 10.,
            "seed":None,            # random seed used for the generative model.
            "positive_input": True,
            "positive_phi": False,
            "REPORT": False,
            }

PATH_FIGURES = "./figures/spiking_based/"
PATH_CACHING = cmn.HOME+"/data/LCA/tmp/"
PATH_RESULTS = cmn.HOME+"/data/LCA/spiking3/"
cmn.mkdir_p(PATH_FIGURES)
cmn.mkdir_p(PATH_RESULTS)
```

## The input factor sweep

### k = 4

```python
file_name0 = PATH_CACHING + "/spike-based-example3_0.pkl"
```

```python
%%cache $file_name0 input_factors0, errors0, nr_active0, pars0

window = 500
pars0=default_parameters.copy()
pars0["seed"]=12345678
pars0["t_end"] = 1000
pars0["k"] = 4
input_factors0 = [0.1, 3.5, 4., 4.5, 4.6, 4.7, 4.8, 4.9, 5., 6., 7., 8., 9., 10., 11., 12., 14., 16., 18., 20., 24., 28., 32., 36., 40., 50., 60., 70., 80., 90., 100.]
nr_active0, errors0 = [], []
for inp_f in input_factors0:
    pars0["input_factor"]=inp_f
    SMs, M, GM = simulate(pars0)    
    errors0.append(get_dynamic_error(get_dynamic_a(M, window, 1, pars0["t_end"])[:,window/2:pars0["t_end"]-window/2], GM)[-1])
    nr_active0.append(number_of_active_neurons(M, window/2.))

results = {"inp_f": input_factors0,
           "err": errors0,
           "nr_act": nr_active0,
           "pars":pars0,
            }
cmn.save_pickle(PATH_RESULTS+"/k=4.pkl", results)
```

```python
plot_reconstruction_error(input_factors0, nr_active0, errors0, pars0["Ndic"])
savefig(PATH_FIGURES + "spike-reconstruction-error3_0.png")
```

## APPENDIX: Calculating the mean error if you take a random direction vector and then optimize it as I do for spiking model

```python
err = 0
N = 18
NN = 10000
for i in xrange(NN):
    f = np.random.randn(N)
    f = f/np.linalg.norm(f)
    y = np.random.randn(N)
    y = y/np.linalg.norm(y)
    rec_op = np.dot(y, f) * y
    np.linalg.norm(y)
    #print np.dot(y, f), np.linalg.norm(f - rec_op), np.linalg.norm(rec_op)
    err += np.linalg.norm(f - rec_op)
print err/NN
```
