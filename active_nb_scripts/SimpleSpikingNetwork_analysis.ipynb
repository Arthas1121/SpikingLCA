## Things I learned:

- The b_w images coming from .png files seem to give [0, 1] input, where white is 1.
- You can add constant to the nodes by using the ANN layer and Vshift, but VThreshold can not be -Infinity (ideally just a bit bigger)
- Foreground parameter in Retina will set the provisional firing rate when the input is 1. I say provisional as the actual firing rate is probably smaller if you have refractory peroid set and something else which I can not remember now ... Foreground **will not** be the maximum firing rate, that is rather enforced by dt as you can only have one spike per timestep.
- Retina will be a layer for transfering from rates to spikes. Not sure if this is the only such layer, or just an implementation of a more general layer which does this.
- All the times or steps parameters in the params files should correspond to the global time (and not to the actual step). If not it's a bug.
- LIFGap Layer probe can be used to record the conductances. I haven't tested it yet, though.
- It makes sense to use the SpikeTrain and SpikePopulation.
- I still don't know what burst parameters are doing in Retina.
- GapLayer is a Clone layer. It is needed when you need both the v and a from a certain layer. In Gap case for Gap Junctions.
