# A rate-based example of the LCA dynamics:

The big question here is - does the Casper example works in the rate model? We know it fails in the spiking model (except for the trivial case of the Identity Dictionary matrix). 

### Things I notice:

* It works with the trivial identity dictionary.
* If I use everything positive, including the threholds, I get into this corrupt region, where I either can only get a small portion of the neurons to be active, or if I artificially decrease the recurrent network strength (which is all inhibitory), I get more neurons active, but the rec error is still big!!! Important finding, as this pertains to our neuronal network.
* What happens with only positive dict. elements and only pos. threhoslds is that we get the same effect as if we had less dict. elements (smaller overcompletness), as for any given image/input, half of the dict. are out by default. In fact the artificial reduction of the overcompletness would happen for any positive-only threshold! Why? Well, again for each input image, you can only use the dictionary elements which are positivelly correlated with the image, which at random means using about half. Could this be improved by having a better dictionary?

```python
%load_ext ipycache
%load_ext autoreload
%autoreload 2

import neurovivo.plotting.plotting as plt
import matplotlib.pylab as pyl
import numpy as np
import neurovivo.common as cmn
import brian as br

import LCA.LCA_common as LCAcmn
from LCA.LCA_simulate import simulate_rate5 as simulate
from LCA.LCA_plotting import plot_a_evolutions, plot_reconstruction_error, visualize_vector_comparison
from LCA.LCA_analysis import get_dynamic_error, number_of_active_neurons, reconstruction_evaluation, opt_error, error_of_normalized

default_parameters={
            "Ndic":64,              # number of dictionaries
            "Ndim":16,              # input dimension (smaller or equal then Ndic)
            "k":4,                  # number of dictionaries used to create the input
            "tau_m":1,              # in ms
            "dt": 0.1,
            "t_end": 1000,
            "threshold": 0.4,
            "thr_func": None,       # The T(.) function taking two parameters (lambda and u)
            "input_factor":1.,
            "thr_coeff": 1.,
            "GM":None,
            }

PATH_FIGURES = cmn.HOME+"/ws/LCA/figures/rate_based6/"
PATH_CACHING = cmn.HOME+"/data/LCA/tmp/"
PATH_RESULTS = cmn.HOME+"/data/LCA/rate6/"
cmn.mkdir_p(PATH_FIGURES)
cmn.mkdir_p(PATH_RESULTS)

def soft_threshold_pos(u, thresh):
    vals = u.copy()                                                                   
    vals[(vals<thresh)]=0.
    vals[vals>=thresh] -= thresh
    return vals

def hard_threshold_sim(u, thresh):
    vals = u.copy()                                                                   
    vals[(vals<thresh) & (vals>-thresh)]=0.
    return vals

def hard_threshold_pos(u, thresh):
    vals = u.copy()                                                                   
    vals[(vals<thresh)]=0.
    return vals
```

```python
casper = np.load(cmn.HOME+"/ws/LCA/tmp/casper.npy")
casper[8,13]=125
casper=casper-np.min(casper)
casper = -1*casper+np.max(casper)
casper=LCAcmn.normalize(casper)
casper1d = casper.reshape(1, 19*19)
```

```python
from matplotlib import cm
pyl.figure(figsize=(12,4))
pyl.subplot(1,2,1)
pyl.imshow(casper, interpolation="nearest", cmap=cm.Greys)
pyl.subplot(1,2,2)
pyl.hist(casper1d[0]);
```

## With the identity matrix - it works

```python
Ndic = 361
Ndim = 361
k = 16
seed = 1234567
GM = LCAcmn.generative_model_data(Ndic, Ndim, k, positive_phi=True, positive_input=True, identity_dic=True, seed=seed)
GM["x"] = casper1d[0]
GM["i_stim"] = np.dot(GM["phi"].T, GM["x"])
GM["a"] = None

pars=default_parameters.copy()
pars["t_end"] = 3000
pars["Ndic"] = Ndic
pars["Ndim"] = Ndim
pars["GM"] = GM
pars["thresh_func"] = soft_threshold_pos
pars["k"] = k
pars["input_factor"]=10.
pars["threshold"] = 0.02
window = pars["t_end"]/8.

SMs_a = simulate(pars)    
#SMs_a = np.array([SMs[3][i][:] for i in xrange(pars["Ndic"])])
rec = np.dot(GM["phi"], SMs_a)
rec_of_mean_a = np.dot(GM["phi"], np.mean(SMs_a, 1))

print get_dynamic_error(SMs_a, GM)[-1], error_of_normalized(rec.T[-1],GM["x"]), error_of_normalized(rec_of_mean_a,GM["x"])
print len(SMs_a.T[-1][SMs_a.T[-1]>0.001])
print np.sum(SMs_a.T[-1])
```

```python
visualize_vector_comparison(rec.T[5::300], GM["x"], position="vertical", scale=.2)
pyl.figure(figsize=(12,4))
end_as = np.mean(SMs_a.T[-500:], 0)
pyl.subplot(1,2,1)
pyl.hist(end_as, 25);
#pyl.ylim(0,10)
pyl.subplot(1,2,2)
#br.raster_plot(M)
pyl.figure(figsize(12,4))
pyl.subplot(1,2,1)
imshow(np.reshape([GM["x"]], (19, 19)), interpolation="nearest", cmap = cm.Greys)
pyl.subplot(1,2,2)
imshow(np.reshape([rec.T[-1]], (19, 19)), interpolation="nearest", cmap = cm.Greys)
```
