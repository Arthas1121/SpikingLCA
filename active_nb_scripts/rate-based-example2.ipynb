# A rate based example of the LCA dynamics - this time with the surrogate dictionaries:

Same as rate-based-example, but now I don't create a random G, but rather random dictionaries, which I use to construct G.

### Things I notice:

* You can control the sparsness with the threshold value.
* The size of the reconstructions drops (along with the error) as you increase the threhold.
* Normalizing the reconstructions does not decreaese the error - the opposite.

```python
%load_ext ipycache

import neurovivo.plotting.plotting as plt
import numpy as np
import common as cmn
import brian as br

np.random.seed(seed=123456789)
Ndic = 64
Ndim = 16

# create N normalized dictionaries - nonoptimized
b = np.random.randn(Ndic,Ndim)
norms = np.apply_along_axis(np.linalg.norm, 1, b)
b_n = b / norms.reshape(-1,1)

# create a G matrix from the normalized dictionaries
G = np.zeros((Ndic,Ndic))
for i in xrange(Ndic):
    for j in xrange(i,Ndic):
        G[i][j] = np.dot(b_n[i], b_n[j])
G = G + G.T - 2*np.identity(Ndic)
G[np.abs(G)<0.00000000001] = 0.

# create i_stim based on the random input vector we are sparsifyign

x = np.random.randn(Ndim)
x_norm = 1.*x/np.linalg.norm(x)
i_stim = np.dot(b_n,x_norm)
    
default_parameters={
            "N":Ndic,
            "v_reset":-60.,         #in mV
            "v_thr":-50.,           #in mV
            "tau_m":20,             #in ms
            "v_rest":-60.,          #in mV
            "dt_delta_thr":10,      #in ms
            "v_rev_e":0.,           #in mV
            "v_rev_i":-80.,         #in mV
            "tau_e":5,              #in ms
            "tau_i":10,             #in ms - note theory requires formally tau_i < tau_e
            "t_refr":5,             #in ms
            "G":G,
            "i_s":i_stim,
            "tau": 1,
            "dt": 0.1,
            "t_end": 1000,
            "threshold": 0.4,
            "REPORT": False,
            }

PATH_FIGURES = cmn.HOME + "/data/LCA/figures/"
PATH_FIGURES = "./figures/"
cmn.mkdir_p(PATH_FIGURES)

def sparsness(vec):
    return 1-1.*np.count_nonzero(vec)/len(vec)

def plot_edge(SMs, i):
    figure()
    plot(SMs[0][i], label="u")
    plot(SMs[1][i], label="a")
    plot(SMs[2][i], label="li")
    legend()

def plot_a_evolutions(SMs, N):
    [plot(SMs[1][i], label=str(i), color = "blue") for i in xrange(N)]
    legend()
    
def plot_u_evolutions(SMs, N):
    [plot(SMs[0][i], label=str(i), color = "red") for i in xrange(N)]
    legend()
    
def reconstruction(b_n, SMs, normalize=False):
    rec = np.dot(b_n.T,SMs[1][:][:,-1])
    if np.linalg.norm(rec) != 0 and normalize:
        rec = rec/np.linalg.norm(rec)    
    return rec

def reconstruction_error(b_n, x, SMs, normalize_reconstruction=False):
    rec = reconstruction(b_n, SMs, normalize=normalize_reconstruction)
    return np.linalg.norm(x - rec)
```

```python
def simulate_rate(parameters=default_parameters):
    x = cmn.Bunch(parameters)
    tau_m_inv = 1./(x.tau*br.ms)

    br.clear(all=True) # clears away all the brian objects from previous simulations.
    
    def thresh_func(values, threshold):
        vals = values.copy()
        vals[(vals<threshold) & (vals>-threshold)]=0.
        return vals
    
    eqs = '''
          du/dt=(-u-li+i_s)*tau_m_inv     :br.Hz
          a                               :br.Hz
          i_s                             :br.Hz
          li                              :br.Hz
          '''
    
    simulation_clock = br.Clock(dt=x.dt*br.ms)
    reporting_clock = br.Clock(dt=1000*br.ms)
    recording_clock = br.Clock(dt=1*br.ms)
    
    P = br.NeuronGroup(x.N, model=eqs, threshold=200000*br.Hz, clock=simulation_clock)
    P.i_s = x.i_s
    P.a = 0
    
    @br.network_operation(clock=simulation_clock, when="end")
    def set_input(simulation_clock):
        P.li = np.dot(x.G, P.a)
        P.a = thresh_func(P.u, x.threshold)
        
    if x.REPORT:
        @br.network_operation(clock=reporting_clock, when="start")
        def report(reporting_clock):
            print reporting_clock.t
    
    SMs = []
    SM_1 = br.StateMonitor(P, "u", record=True, timestep=1, clock=recording_clock)
    SM_2 = br.StateMonitor(P, "a", record=True, timestep=1, clock=recording_clock)
    SM_3 = br.StateMonitor(P, "li", record=True, timestep=1, clock=recording_clock)
    SM_4 = br.StateMonitor(P, "i_s", record=True, timestep=1, clock=recording_clock)
    SMs.append(SM_1)
    SMs.append(SM_2)
    SMs.append(SM_3)
    SMs.append(SM_4)
    
    if x.REPORT:
        network = br.Network(P, set_input, report, *SMs)
    else:
        network = br.Network(P, set_input, *SMs)
        
    # starting the simulation
    network.run(x.t_end*br.ms)

    if x.REPORT:
        print "simulation finished"
    return SMs
```

## Threshold dependence

### Dependence of the sparsity on the threshold

```python
%%cache /Users/mpelko/data/LCA/tmp/rate-based-example2.pkl reconstructions, sparsnesses, error, error_norm, reconstruction_norms, reconstruction_norms_norm

# Warning: takes about 5 minutes on Mac Air
pars = default_parameters.copy()
pars["t_end"] = 2000
thresholds = np.linspace(0.,.7,20)
reconstructions, sparsnesses, error, error_norm, reconstruction_norms, reconstruction_norms_norm = [],[],[],[],[],[]
for t in thresholds:
    pars["threshold"] = t
    SMs = simulate_rate(pars)
    sparsnesses.append(sparsness(SMs[1][:][:,-1]))
    error.append(reconstruction_error(b_n, x_norm, SMs))
    reconstruction_norms.append(np.linalg.norm(reconstruction(b_n,SMs)))
    error_norm.append(reconstruction_error(b_n, x_norm, SMs, normalize_reconstruction=True))
    reconstruction_norms_norm.append(np.linalg.norm(reconstruction(b_n,SMs, normalize=True)))
    reconstructions.append(reconstruction(b_n,SMs))
```

```python
plot(thresholds, sparsnesses)
xlabel(r"threshold ($\lambda$)")
ylabel("sparsness")
savefig(PATH_FIGURES + "sparsness_on_threshold.png")
```

### How good is the reconstruction (dependence on the threshold and sparsity)?

```python
def plot_reconstrustion_error(thresholds, sparsnesses, reconstruction_norms, error):
    figure(figsize=(12,4))
    subplot(1,2,1)
    plot(thresholds, error, label=r"$\|x-\phi a\|$")
    xlabel(r"threshold ($\lambda$)")
    plot(thresholds, reconstruction_norms, label=r"$\|\phi a\|$")
    legend(loc="best")
    subplot(1,2,2)
    plot(sparsnesses,error, label=r"$\|x-\phi a\|$")
    xlabel("sparsness")
    plot(sparsnesses, reconstruction_norms, label=r"$\|\phi a\|$")
    legend(loc="best")

plot_reconstrustion_error(thresholds, sparsnesses, reconstruction_norms, error)
savefig(PATH_FIGURES + "rate-reconstruction-error.png")
```

### What happens if we normalize the reconstructions? Would that be better or worse in terms of errors?

```python
plot_reconstrustion_error(thresholds, sparsnesses, reconstruction_norms_norm, error_norm)
```

```python
plot(error, error_norm)
xlabel("error")
ylabel("error of normalized reconstruction")
plot([0,1],[0,1])
```

The result is worse. It seems there is a reason why the norm gets smaller.

## APPENDIX 1: Calculating the error if you take a random normalized vector as a solution

```python
err = 0
N=64
NN = 10000
for i in xrange(NN):
    f = np.random.randn(N)
    f = f/np.linalg.norm(f)
    y = np.random.randn(N)
    y = y/np.linalg.norm(y)
    err += np.linalg.norm(f - y)
print err/NN
```

Turns out it's $\sqrt 2$ (seems independent of the dimensions). This can be probably shown easily analytically.
