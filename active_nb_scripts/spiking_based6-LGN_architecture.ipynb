# A spike-based example of the LCA dynamics - in LGN architecture:

In this spike-based implementation we use:

* **The architecture with the residual/error layer (LGN)**
* The generative model for creating the inputs and dictionaries,
* We only use the positive dictionary elements and inputs
* We implement stochastic external input.

### Things I notice:

* It seems more difficult to get the spiking LCA running in the LGN architecture
* The less noisy the input (higher frequency at smaller amplitude), the better the reconstruction.

```python
%load_ext autoreload
%autoreload 2

import matplotlib.pylab as pyl
import numpy as np
import brian as br

import LCA.LCA_common as cmn
#from LCA.LCA_simulate import simulate_spiking5 as simulate
from LCA.LCA_plotting import plot_a_evolutions, plot_reconstruction_error, visualize_vector_comparison
from LCA.LCA_analysis import get_dynamic_error, number_of_active_neurons, reconstruction_evaluation, opt_error, error_of_normalized

default_parameters={
            "Ndic":18,              # number of dictionaries
            "Ndim":12,              # input dimension (smaller or equal then Ndic)
            "k":4,                  # number of dictionaries used to create the input
            "v_reset":-60.,         #in mV
            "v_thr_V1":-55.,        #in mV
            "v_thr_LGN":-58.,       #in mV
            "tau_m":20,             #in ms
            "v_rest":-60.,          #in mV
            "v_rev_e":0.,           #in mV
            "v_rev_i":-80.,         #in mV
            "w_e":0.3/3.,
            "w_i":0.3,            
            "w_s":0.1,            
            "tau_e":5,              #in ms
            "tau_i":5,              #in ms
            "tau_a":250,            #in ms
            "t_refr":.25,           #in ms
            "dt": 0.1,
            "t_end": 1000,
            "tonic_input": False,
            "input_factor_t": 10.,
            "input_factor_s": 10.,
            "GM": None,
            "positive_input": True,
            "positive_phi": True,
            "REPORT": False,
            }

PATH_FIGURES = cmn.HOME+"/ws/LCA/figures/spiking_based6/"
PATH_CACHING = cmn.HOME+"/data/LCA/tmp/"
PATH_RESULTS = cmn.HOME+"/data/LCA/spiking6/"
cmn.mkdir_p(PATH_FIGURES)
cmn.mkdir_p(PATH_RESULTS)
```

```python
def simulate_spiking6_2(parameters):

    x = cmn.Bunch(parameters)
    tau_m_inv = 1./(x.tau_m*br.ms)
    tau_e_inv = 1./(x.tau_e*br.ms)
    tau_i_inv = 1./(x.tau_i*br.ms)
    tau_a_inv = 1./(x.tau_a*br.ms)

    assert x.GM, "generate the surrogate data and pass it into the fuction parameters."

    GM = x.GM
    br.clear(all=True) # clears away all the brian objects from previous simulations.

    eqs_LGN = '''
              dvm/dt=((x.v_rest*br.mV-vm) + ge*(x.v_rev_e*br.mV-vm) + gi*(x.v_rev_i*br.mV-vm) + i_s)*tau_m_inv :br.mV
              dge/dt = -ge*tau_e_inv :1
              dgi/dt = -gi*tau_i_inv :1
              i_s                 :br.mV
              '''

    eqs_V1 =  '''
              dvm/dt=((x.v_rest*br.mV-vm) + ge*(x.v_rev_e*br.mV-vm))*tau_m_inv :br.mV
              dge/dt = -ge*tau_e_inv :1
              da/dt = -a*tau_a_inv   :br.Hz
              '''
    
    simulation_clock = br.Clock(dt=x.dt*br.ms)
    recording_clock = br.Clock(dt=1*br.ms)

    P_LGN = br.NeuronGroup(x.Ndim, model=eqs_LGN, threshold=x.v_thr_LGN*br.mV, reset=x.v_reset*br.mV, refractory=x.t_refr*br.ms, clock=simulation_clock)
    P_V1 = br.NeuronGroup(x.Ndic, model=eqs_V1, threshold=x.v_thr_V1*br.mV, reset=x.v_reset*br.mV, refractory=x.t_refr*br.ms, clock=simulation_clock)
    P_LGN.vm =  (((np.random.random((x.Ndim))-0.5) * 0.5 * (x.v_rest-x.v_thr_LGN)) + x.v_rest) * br.mV
    P_V1.vm =  (((np.random.random((x.Ndic))-0.5) * 0.5 * (x.v_rest-x.v_thr_V1)) + x.v_rest) * br.mV
    
    if x.tonic_input:
        P_LGN.i_s = x.input_factor_t * GM["i_stim"] * br.mV
    P_V1.a = 0 

    W_V1_to_LGN = GM['phi'].T
    W_LGN_to_V1 = GM['phi']
                
    V1_to_LGN = br.Connection(P_V1, P_LGN, 'gi')
    V1_to_LGN.connect(P_V1, P_LGN, x.w_i*W_V1_to_LGN)
    LGN_to_V1 = br.Connection(P_LGN, P_V1, 'ge')
    LGN_to_V1.connect(P_LGN, P_V1, x.w_e*W_LGN_to_V1)
        
    Ca = br.IdentityConnection(P_V1, P_V1, 'a', weight=tau_a_inv)
    V1_to_V1 = br.IdentityConnection(P_V1, P_V1, 'ge', weight=x.w_e)
        
    if not x.tonic_input:
        Ps = br.PoissonGroup(x.Ndim, rates = GM["x"]*x.input_factor_s*br.Hz, clock=simulation_clock)
        Cs = br.IdentityConnection(Ps, P_LGN, 'ge', weight=x.w_s)        
            
    M_LGN = br.SpikeMonitor(P_LGN)
    M_V1 = br.SpikeMonitor(P_V1)
        
    SMs = []
    SMs.append(br.StateMonitor(P_LGN, "vm", record=True, timestep=1, clock=recording_clock))
    SMs.append(br.StateMonitor(P_LGN, "ge", record=True, timestep=1, clock=recording_clock))
    SMs.append(br.StateMonitor(P_LGN, "gi", record=True, timestep=1, clock=recording_clock))
    SMs.append(br.StateMonitor(P_V1, "vm", record=True, timestep=1, clock=recording_clock))
    SMs.append(br.StateMonitor(P_V1, "ge", record=True, timestep=1, clock=recording_clock))
#    SMs.append(br.StateMonitor(P_V1, "gi", record=True, timestep=1, clock=recording_clock))
    SMs.append(br.StateMonitor(P_V1, "a", record=True, timestep=1, clock=recording_clock))
    SMs.append(br.StateMonitor(P_LGN, "i_s", record=True, timestep=1, clock=recording_clock))
    
    if x.tonic_input: 
        network = br.Network(P_LGN, P_V1, V1_to_LGN, LGN_to_V1, V1_to_V1, Ca, M_LGN, M_V1, *SMs)
    else:
        network = br.Network(P_LGN, P_V1, Ps, V1_to_LGN, LGN_to_V1, V1_to_V1, Ca, Cs, M_LGN, M_V1, *SMs)
            
    # starting the simulation
    network.run(x.t_end*br.ms)

    return SMs, M_LGN, M_V1
```

```python
def spiking_rates(M, total_time):
    return 1000.*np.array([len(M.spiketimes[i]) for i in xrange(len(M.spiketimes.keys()))])/total_time
        
def pop_rates(M, dt, total_time):
    nbin = int(1.*total_time/dt)
    return 1000.*np.histogram(M.it[1], np.linspace(0, total_time/1000., nbin))[0]/nbin
```

```python
Ndic = 256
Ndim = 64
k = 2
seed = 1234567
GM = LCAcmn.generative_model_data(Ndic, Ndim, k, positive_phi=True, positive_input=True, seed=seed)
```

```python
pars=default_parameters.copy()
pars["t_end"] = 3000
pars["Ndic"] = Ndic
pars["Ndim"] = Ndim
pars["v_thr_LGN"] = -58.
pars["GM"] = GM
pars["k"] = k
pars["w_e"] = 8*0.1/3./Ndim
pars["w_i"] = 1200.*.6125/Ndic
pars["w_s"] = 1.*.006125
pars["tau_e"] = 40
pars["tau_i"] = 80
pars["v_thr"] = -57
pars["tonic_input"] = False
pars["input_factor_s"]=1280./k

window = pars["t_end"]/8.

SMs, M_LGN, M_V1 = simulate_spiking6_2(pars)    
SMs_a = np.array([SMs[5][i][:] for i in xrange(pars["Ndic"])])
rec = np.dot(GM["phi"], SMs_a)
rec_of_mean_a = np.dot(GM["phi"], np.mean(SMs_a, 1))

print get_dynamic_error(SMs_a, GM)[-1], error_of_normalized(rec.T[-1],GM["x"]), error_of_normalized(rec_of_mean_a,GM["x"])
print number_of_active_neurons(M_V1, window/2.)
print np.sum(SMs_a.T[-1])
```
