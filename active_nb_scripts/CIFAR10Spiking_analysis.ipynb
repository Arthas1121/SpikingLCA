# Putting together the analysis and visualizations for the CIFAR10Spiking LCA trial


```python
%load_ext autoreload
%autoreload 2

import LCA.LCA_common as cmn
import LCA.LCA_octave_io as oio
import neurovivo.plotting.plotting as nvplt
import matplotlib as mpl
mpl.rc("text", usetex=False)
import matplotlib.pyplot as pyl
import matplotlib.cm as cm
from matplotlib import animation
import LCA.LCA_plotting as plt
from LCA.LCA_analysis import sparse_to_full, reconstruct_image, dict_set_at_idx, spike_population_from_pvp
import neurovivo.common as nvcmn

PATH_OUTPUT = cmn.HOME+"/data/CIFAR10Spiking/"

PATH_FIGURES = cmn.HOME + "/data/figures/CIFAR10Spiking/"
cmn.mkdir_p(PATH_FIGURES)
```

## ACTIVITIES

```python
[gang_data, gang_header] = oio.readpvpfile(PATH_OUTPUT + "/a1_ganglion.pvp")
[retina_on_data, retina_on_header] = oio.readpvpfile(PATH_OUTPUT + "/a2_retina_on.pvp")
[retina_off_data, retina_off_header] = oio.readpvpfile(PATH_OUTPUT + "/a3_retina_off.pvp")
[residual_on_data, residual_on_header] = oio.readpvpfile(PATH_OUTPUT + "/a4_residual_on.pvp")
[residual_off_data, residual_off_header] = oio.readpvpfile(PATH_OUTPUT + "/a5_residual_off.pvp")
[V1_on_data, V1_on_header] = oio.readpvpfile(PATH_OUTPUT + "/a6_V1_on.pvp")
[V1_off_data, V1_off_header] = oio.readpvpfile(PATH_OUTPUT + "/a7_V1_off.pvp")
[reconstruction_data, reconstruction_header] = oio.readpvpfile(PATH_OUTPUT + "/a8_LIF_reconstruction.pvp")
```

### From Ganglion layer (rate) to retina (spikes)

```python
i = 1
display_t = 500
ganglion_image = gang_data[i][0][1]
retina_on_sp = spike_population_from_pvp(retina_on_data, retina_on_header, 0.1).population_part(i*display_t,(i+1)*display_t)
retina_off_sp = spike_population_from_pvp(retina_off_data, retina_off_header, 0.1).population_part(i*display_t,(i+1)*display_t)
print "The full retina fires at the total of {} Hz.".format(1000*(np.sum(retina_on_sp.rates())+np.sum(retina_off_sp.rates()))/display_t)
```

```python
pyl.figure(figsize=(18,4))
pyl.subplot(1,4,1)
pyl.title("ganglion")
pyl.imshow(ganglion_image.T, cmap=cm.Greys)
pyl.subplot(1,4,2)
pyl.title("retina_on")
pyl.imshow(np.reshape(retina_on_sp.rates(), (32,32)), cmap=cm.Greys)
pyl.subplot(1,4,3)
pyl.title("retina_off")
pyl.imshow(np.reshape(retina_off_sp.rates(), (32,32)), cmap=cm.Greys)
pyl.subplot(1,4,4)
pyl.title("retina_combined")
pyl.imshow(np.reshape(np.array(retina_on_sp.rates())-np.array(retina_off_sp.rates()), (32,32)), cmap=cm.Greys);
```

```python
pyl.figure(figsize=(18,4))
pyl.subplot(1,3,1)
pyl.title("retina_on")
pyl.hist(retina_on_sp.rates())
pyl.subplot(1,3,2)
pyl.title("retina_off")
pyl.hist(retina_off_sp.rates())
pyl.subplot(1,3,3)
pyl.title("retina_both")
pyl.hist(retina_off_sp.rates()+retina_on_sp.rates());
```

```python
pyl.figure(figsize=(18,4))
pyl.subplot(1,2,1)
pyl.title("retina_on")
nvplt.plot_spike_population(retina_on_sp)
pyl.subplot(1,2,2)
pyl.title("retina_off")
nvplt.plot_spike_population(retina_off_sp)
```
