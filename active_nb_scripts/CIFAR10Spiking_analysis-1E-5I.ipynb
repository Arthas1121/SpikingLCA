```python
dict_pos = dict_from_flat_data(weights_pos_data, weights_pos_header)
dict_neg = dict_from_flat_data(weights_neg_data, weights_neg_header)
V1_on_act=np.rollaxis(np.array(V1_on_sp.rates()).reshape(8,8,32),1)
V1_off_act=np.rollaxis(np.array(V1_off_sp.rates()).reshape(8,8,32),1)
```

```python
def reconstruct_image_from_activities(V1_on, V1_off, dict_pos, dict_neg):
    image1 = reconstruct_image(V1_on, dict_pos, nx_rel_scale=.25, ny_rel_scale=.25)
    image2 = reconstruct_image(V1_on, dict_neg, nx_rel_scale=.25, ny_rel_scale=.25)
    image3 = reconstruct_image(V1_off, dict_pos, nx_rel_scale=.25, ny_rel_scale=.25)
    image4 = reconstruct_image(V1_off, dict_neg, nx_rel_scale=.25, ny_rel_scale=.25)
    return image1, image2, image3, image4, image1+image4-image2-image3
```

```python
im1, im2, im3, im4, im5 = reconstruct_image_from_activities(V1_on_act, V1_off_act, dict_pos, dict_neg)
```
