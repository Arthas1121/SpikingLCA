# A spike-based example of the LCA dynamics:

In this spike-based implementation we use:

* The generative model for creating the inputs and dictionaries,
* We only use the positive dictionary elements,
* We implement **stochastic external input**.

### Things I notice:

* I get into difficulties with the variabilities. When I generate the input by using many dictionary elements (k>>), the variability of the input goes down dramatically.  The result is, that the time variance of the individual neuron excitatory input (due to the stochastic input spiking) might exceede the the variability of the mean of the individual inputs. 
    * This might be improved by having more optimal dictionary. The dictionaries are learned under the confirmed assumption of sparsness od the data (confirmed in the case of vision). This implies k values are realtivelly smaller and hopefully a nice variability in the input.
* The casper example doesn't really work even for the tonic excitatory input. The only case where it works is the obvious setting the dictionary to the identity. I am currently blaming the random dictionary selection, but it might be other things. I should check if the example with random dictionaries works on the rate model!
* Well, the previous point is not *quite* true. I even manage to get the casper going for the spiking input 
    * If I have enough dictionaries I can use and the input.
    * If my input spiking has a really high frequency (see figures below).

### Things to look into

* The effect of noise in the rate model.

```python
%load_ext ipycache
%load_ext autoreload
%autoreload 2

import neurovivo.plotting.plotting as plt
import matplotlib.pylab as pyl
import numpy as np
import neurovivo.common as cmn
import brian as br

import LCA.LCA_common as LCAcmn
from LCA.LCA_simulate import simulate_spiking5 as simulate
from LCA.LCA_plotting import plot_a_evolutions, plot_reconstruction_error, visualize_vector_comparison
from LCA.LCA_analysis import get_dynamic_error, number_of_active_neurons, reconstruction_evaluation, opt_error, error_of_normalized

default_parameters={
            "Ndic":18,              # number of dictionaries
            "Ndim":12,              # input dimension (smaller or equal then Ndic)
            "k":4,                  # number of dictionaries used to create the input
            "v_reset":-60.,         #in mV
            "v_thr":-55.,           #in mV
            "tau_m":20,              #in ms
            "v_rest":-60.,          #in mV
            "v_rev_e":0.,           #in mV
            "v_rev_i":-80.,         #in mV
            "w_e":0.3/3.,
            "w_i":0.3,            
            "w_s":0.1,            
            "tau_e":5,              #in ms
            "tau_i":5,              #in ms
            "tau_a":250,            #in ms
            "t_refr":.25,             #in ms
            "dt": 0.1,
            "t_end": 1000,
            "tonic_input": False,
            "input_factor_t": 10.,
            "input_factor_s": 10.,
            "GM": None,
            "positive_input": True,
            "positive_phi": True,
            "REPORT": False,
            }

PATH_FIGURES = cmn.HOME+"/ws/LCA/figures/spiking_based5/"
PATH_CACHING = cmn.HOME+"/data/LCA/tmp/"
PATH_RESULTS = cmn.HOME+"/data/LCA/spiking5/"
cmn.mkdir_p(PATH_FIGURES)
cmn.mkdir_p(PATH_RESULTS)
```

```python
Ndic = 128
Ndim = 16
k = 16
seed = 1234567
GM = LCAcmn.generative_model_data(Ndic, Ndim, k, positive_phi=True, positive_input=True, seed=seed)

pars=default_parameters.copy()
pars["t_end"] = 3000
pars["Ndic"] = Ndic
pars["Ndim"] = Ndim
pars["GM"] = GM
pars["k"] = k
pars["w_i"] = .25*.06125
pars["w_s"] = 0.5*.006125
pars["tau_e"] = 40
pars["tau_i"] = 80
pars["v_thr"] = -57
pars["tonic_input"] = False

pars["input_factor_s"]=64*128./k

window = pars["t_end"]/8.

SMs, M, GM = simulate(pars)    
SMs_a = np.array([SMs[3][i][:] for i in xrange(pars["Ndic"])])
rec = np.dot(GM["phi"], SMs_a)
rec_of_mean_a = np.dot(GM["phi"], np.mean(SMs_a, 1))

print get_dynamic_error(SMs_a, GM)[-1], error_of_normalized(rec.T[-1],GM["x"]), error_of_normalized(rec_of_mean_a,GM["x"])
print number_of_active_neurons(M, window/2.)
print np.sum(SMs_a.T[-1])
```

```python
visualize_vector_comparison(rec.T[5::300], GM["x"], position="vertical", scale=1.)
visualize_vector_comparison(SMs_a.T[10::300], GM["a"], position="vertical", scale=.5)
visualize_vector_comparison(np.array([np.mean(SMs_a, 1)]), GM["a"], position="vertical", scale=.5)
pyl.figure(figsize=(12,4))
end_as = np.mean(SMs_a.T[-500:], 0)
pyl.subplot(1,2,1)
pyl.hist(end_as, 25);
pyl.ylim(0,10)
pyl.subplot(1,2,2)
br.raster_plot(M)
```

```python
vms = [SMs[0][i][:] for i in xrange(Ndic)]
ges = [SMs[1][i][:] for i in xrange(Ndic)]
gis = [SMs[2][i][:] for i in xrange(Ndic)]
pyl.figure(figsize=(16,4))
pyl.subplot(1,3,1)
plot(np.array(ges[10:15]).T, linewidth=.5)
#pyl.ylim(-0.05,0.55)
pyl.subplot(1,3,2)
plot(np.array(gis[10:15]).T, linewidth=0.5);
#pyl.ylim(0.,0.55)
pyl.subplot(1,3,3)
ns = pyl.hist(GM["i_stim"]*pars["input_factor_s"], );
xlim(0, 1.5*np.max(ns[1]));
```

```python
pyl.figure(figsize=(12,4))
pyl.subplot(1,2,1)
pyl.plot(get_dynamic_error(SMs_a, GM), label=r"$\||\phi a|-|x|\|$")
pyl.plot([np.linalg.norm(LCAcmn.normalize(a) - LCAcmn.normalize(SMs_a.T[-1])) for a in SMs_a.T], label=r"$\||a|-|a_{end}|\|$")
leg = pyl.legend()
leg.draw_frame(False)
#pyl.xlim(0,200)
#ylim(0,0.05)
pyl.subplot(1,2,2)
pyl.plot([len(np.nonzero(LCAcmn.normalize(a)[LCAcmn.normalize(a)>0.01])[0]) for a in SMs_a.T], label=r"$\|a\|_0$")
#pyl.xlim(0,200)
leg = pyl.legend()
leg.draw_frame(False)
```
