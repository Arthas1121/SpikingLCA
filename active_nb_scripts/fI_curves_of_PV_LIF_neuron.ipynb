```python
%load_ext autoreload
%autoreload 2
from LCA.LCA_various import run_MeasuringFIcurves, default_MeasuringFICurves_parameters
import LCA.LCA_octave_io as oio
from LCA.LCA_analysis import spike_population_from_pvp
import matplotlib.pyplot as pyl

def get_LIF_curves(params):
    output_path = run_MeasuringFIcurves(params)
    [retina_spike_count, retina_header] = oio.readpvpfile(output_path + "/a2_RetinaSpikeCounter.pvp")
    [LIF_spike_count, LIF_header] = oio.readpvpfile(output_path + "/a4_LIFSpikeCounter.pvp")
    retina_rates = 1000*retina_spike_count[-1][0][1].T.reshape(1,16*16)[0]/stop_time
    LIF_rates = 1000*LIF_spike_count[-1][0][1].T.reshape(1,16*16)[0]/stop_time
    return retina_rates, LIF_rates
```

```python
stop_time = 2000.
params = default_MeasuringFICurves_parameters.copy()
params["stop_time"]=stop_time
ret2LIF_Ws = [0.5, 1., 2., 5., 10.]
rrates = []
lrates = []
for w in ret2LIF_Ws: 
    params["ret2LIF_W"]=w
    rrate, lrate = get_LIF_curves(params)
    rrates.append(rrate)
    lrates.append(lrate)
```

```python
for rrate, lrate, w in zip(rrates, lrates, ret2LIF_Ws):
    plot(rrate, lrate, label="w={}".format(w))
maxx = 1000
pyl.plot([0,maxx], [0,maxx], "k",linestyle="--")
xlim(0,maxx)
ylim(0,400)
leg = pyl.legend(loc="best")
leg.draw_frame(False)
```
