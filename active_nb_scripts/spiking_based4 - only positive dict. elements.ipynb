# A spike-based example of the LCA dynamics:

In this spike-based implementation we use the generative model for creating the inputs and dictionaries and we only use the positive dictionary elements.

### Things I notice:

* $w_i = .5$ seems to work very well.
* As we lower k to very small values, we are better off by increasing $w_i$.
* As we increase the number of the neurons, we need to lower w_i, which makes sense, as inhibitory input to a neurons scales with hte number of neurons as well.
* The scaling in this example is bad. I think the reason is, that as we increase the scale, my dictionary elements are all equaly close to one another (that is, the non-diagonal elemnents of $\phi^T \phi$ are all almost the same). Similarly, any input vector will thus have almost the same direction - all the coordinates will be almost the same. This would probably work much better as we have realisic images and dictionaries. I hope.

### Open questions

* Why do positive dictionarys work and the negative do not?
    * I think it might be that the synchrony plays an important role when we have excitatory connections as well - the neurons is driven by the variability in the input rather then the mean of the input.
    * This might be at least slightly resolved by the stochasticity of the external input?

```python
pars=default_parameters.copy()
pars["seed"]=123456
pars["t_end"] = 300
pars["Ndic"] = 3200
pars["Ndim"] = 1600
pars["k"] = 10
pars["w_i"] = 1.
#pars["positive_phi"]=True
pars["input_factor"]=8.*16/pars["k"]

window = pars["t_end"]/8.

SMs, M, GM = simulate(pars)    
SMs_a = np.array([SMs[3][i][:] for i in xrange(pars["Ndic"])])
rec = np.dot(GM["phi"], SMs_a)

print get_dynamic_error(SMs_a, GM)[-1], error_of_normalized(rec.T[-1],GM["x"])
print number_of_active_neurons(M, window/2.)
print np.sum(SMs_a.T[-1])
```
