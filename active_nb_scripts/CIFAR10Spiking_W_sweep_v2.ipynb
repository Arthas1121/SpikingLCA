# The parameter sweep on the weights strengths for CIFAR10Spiking project
## Version 2 - second trial, more inhibition

Using the v5 template (and v4 params file). I am modifying the strenghts of weights in 3 dimensions (ST, EI, UD). 

* ST: general strenghts of the res->V1->res connections
* EI: The ratio between excitatory and inhibitory str. factor
* UD: The ratio between res->V1 and V1->res str. factor

The weight factor (WF) for a specific connection is calculated as WF = ST * sqrt(EI)^(+-1) * sqrt(UD)^(+-1), where the sign of the exponential depends on the nature of the connection.

V1->V1 connections are scaled with the ST parameter.

## RUNNING THE SIMULATION

```python
def run_simulation_W_sweep(params, ST, EI, UD):
    try:
        params["output_path"]="\"{po}st{ST}_ei{EI}_ud{UD}/\"".format(po=PATH_OUTPUT, ST=ST, EI=EI, UD=UD)
        params["V12res_on_inh_W"] = ST*np.sqrt(1./EI)*np.sqrt(1./UD) 
        params["V12res_on_exc_W"] = ST*np.sqrt(1.*EI)*np.sqrt(1./UD) 
        params["V12res_off_inh_W"] = ST*np.sqrt(1./EI)*np.sqrt(1./UD)  
        params["V12res_off_exc_W"] = ST*np.sqrt(1.*EI)*np.sqrt(1./UD)
        params["res_on2V1_exc_W"] = ST*np.sqrt(1.*EI)*np.sqrt(1.*UD)  
        params["res_on2V1_inh_W"] = ST*np.sqrt(1./EI)*np.sqrt(1.*UD)  
        params["res_off2V1_inh_W"] = ST*np.sqrt(1./EI)*np.sqrt(1.*UD)  
        params["res_off2V1_exc_W"] = ST*np.sqrt(1.*EI)*np.sqrt(1.*UD) 
        params["V12V1_W"] = ST
        run_CIFAR10Spiking(params, version=5)
        return True
    except:
        return False
    
def parallel_run(input_as_list):
    return run_simulation_W_sweep(*input_as_list)

params = default_CIFAR10Spiking_v4_parameters.copy()
params["stop_time"]=10000.

STs = [15, 20, 25]
EIs = [0.25, 0.5, 1.]
UDs = [1., 2., 3.]
```

```python
args = [[params, c[0], c[1], c[2]] for c in list(itertools.product(STs, EIs, UDs))]
pool = multiprocessing.Pool(processes=27)
result = pool.map(parallel_run, args)
print np.all(result), result
```

## Analysis

```python
def get_scs_from_PVP_SC_file(SC_data, idx):
    return SC_data[i+1][0][1] - SC_data[i][0][1]

def reconstruct_image_from_activities(V1_on, dict_pos, dict_neg):
    image1 = reconstruct_image(V1_on, dict_pos, nx_rel_scale=.25, ny_rel_scale=.25)
    image2 = reconstruct_image(V1_on, dict_neg, nx_rel_scale=.25, ny_rel_scale=.25)
    return image1-image2

combos = list(itertools.product(STs, EIs, UDs))

display_period = params["display_period"]

[V12res_w_pos_data, V12res_w_pos_header] = oio.readpvpfile(PATH_V12RES_W_POS)
[V12res_w_neg_data, V12res_w_neg_header] = oio.readpvpfile(PATH_V12RES_W_NEG)
dict_pos = dict_from_flat_data(V12res_w_pos_data, V12res_w_pos_header)
dict_neg = dict_from_flat_data(V12res_w_neg_data, V12res_w_neg_header)

# ganglion activity is the same in all the combos, so I simply take it from the first.
(ST, EI, UD) = combos[0]
combo_folder = "{po}st{ST}_ei{EI}_ud{UD}/".format(po=PATH_OUTPUT, ST=ST, EI=EI, UD=UD)
[gang_data, gang_header] = oio.readpvpfile(combo_folder + "/a1_ganglion.pvp")
[V1_rate_data, V1_rate_header] = oio.readpvpfile(PATH_OUTPUT_RATE + "/V1.pvp")
norm_gang_images=[]
norm_rate_rec_images=[]
std_errors_rate_rec = []
for i in xrange(len(gang_data)):
    ganglion_image = gang_data[i][0][1]
    gan_im = ganglion_image.T
    norm_gang_images.append((gan_im - np.mean(gan_im))/np.std(gan_im - np.mean(gan_im)))
    rate_act = sparse_to_full(V1_rate_data, V1_rate_header, i)
    im3r = reconstruct_image_from_activities(rate_act, dict_pos, dict_neg)
    rate_im = im3r[:,:,0].T;
    norm_rate_rec_images.append((rate_im - np.mean(rate_im))/np.std(rate_im - np.mean(rate_im)))    
    std_errors_rate_rec.append(np.std(norm_rate_rec_images[-1]-norm_gang_images[-1]))
```

```python
res_on_mean_ratess = []
res_off_mean_ratess = []
V1_mean_ratess = []
std_errorss = []
#combos = [(10,1,1)]
for combo in combos:
    print combo
    (ST, EI, UD) = combo
    combo_folder = "{po}st{ST}_ei{EI}_ud{UD}/".format(po=PATH_OUTPUT, ST=ST, EI=EI, UD=UD)
    [V1_SC_data, V1_SC_header] = oio.readpvpfile(combo_folder+"a12_V1_onSpikeCounter.pvp")
    [res_on_SC_data, res_on_SC_header] = oio.readpvpfile(combo_folder+"a10_residual_onSpikeCounter.pvp")
    [res_off_SC_data, res_off_SC_header] = oio.readpvpfile(combo_folder+"a11_residual_offSpikeCounter.pvp")
    res_on_mean_rates = np.zeros(len(V1_SC_data)-1)
    res_off_mean_rates = np.zeros(len(V1_SC_data)-1)
    V1_mean_rates = np.zeros(len(V1_SC_data)-1)
    std_errors = np.zeros(len(V1_SC_data)-1)
    for i in xrange(len(V1_SC_data)-1):
        res_on_mean_rates[i]=1000.*np.sum(get_scs_from_PVP_SC_file(res_on_SC_data, i))/display_period
        res_off_mean_rates[i]=1000.*np.sum(get_scs_from_PVP_SC_file(res_off_SC_data, i))/display_period
        V1_mean_rates[i]=1000.*np.sum(get_scs_from_PVP_SC_file(V1_SC_data, i))/display_period
        V1_rates = 1000.*get_scs_from_PVP_SC_file(V1_SC_data, i)/display_period
        rec_im = reconstruct_image_from_activities(V1_rates, dict_pos, dict_neg)[:,:,0].T
        if abs(np.sum(rec_im))>0:
            norm_rec_im = (rec_im - np.mean(rec_im))/np.std(rec_im - np.mean(rec_im))
        else:
            norm_rec_im = rec_im
        std_errors[i] = np.std(norm_rec_im-norm_gang_images[i])
    res_on_mean_ratess.append(res_on_mean_rates)
    res_off_mean_ratess.append(res_off_mean_rates)
    V1_mean_ratess.append(V1_mean_rates)
    std_errorss.append(std_errors)
```

## Investigating the optimal information transfer

```python
limits = [0.02, 0.1, 0.4, .7]
opt_sparse_spikess = []
opt_sparse_ratess = []
opt_sparse_errorss = []
sparse_opt_spikess = []
for lim in limits:
    opt_sparse_spikes = []
    opt_sparse_rates = []
    opt_sparse_errors = []
    sparse_opt_spikes = []
    for i in xrange(len(norm_gang_images)):
        V1_rates = sparse_to_full(V1_rate_data, V1_rate_header, i)
        V1_rates_tmp = lim*255.*V1_rates/np.max(V1_rates)
        V1_counts_8bit = np.array(V1_rates_tmp, np.uint8)
        print np.max(V1_counts_8bit)
        opt_sparse_spikes.append(np.sum(V1_counts_8bit))
        opt_sparse_rates.append(len(V1_rates.nonzero()[0])*19)
        rec_im_8bit = reconstruct_image_from_activities(V1_counts_8bit, dict_pos, dict_neg)[:,:,0].T
        norm_rec_im_8bit = (rec_im_8bit - np.mean(rec_im_8bit))/np.std(rec_im_8bit - np.mean(rec_im_8bit))
        opt_sparse_errors.append(np.std(norm_gang_images[i]-norm_rec_im_8bit))
    opt_sparse_spikess.append(opt_sparse_spikes)
    opt_sparse_ratess.append(opt_sparse_rates)
    opt_sparse_errorss.append(opt_sparse_errors)
    sparse_opt_spikess.append(sparse_opt_spikes)
```
