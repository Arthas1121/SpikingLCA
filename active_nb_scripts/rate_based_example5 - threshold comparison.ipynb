# A rate based example - checking threshold functions

The rate based example. This time we do a systematic check of several threshold implementations.

### Things I notice:

* See the report 14_02_11
* the positive only threholds are in general worse then the symetric ones,
* the soft threshold performs worse then the other two,
* while mixed and hard threshold seem to be pretty comparable.

```python
%load_ext ipycache
%load_ext autoreload
%autoreload 2

import neurovivo.plotting.plotting as plt
import matplotlib.pyplot as pyl
import numpy as np
import neurovivo.common as cmn

from LCA_common import generative_model_data
from LCA_simulate import simulate_rate5 as simulate
from LCA_plotting import plot_reconstruction_error
from LCA_analysis import reconstruction_evaluation

default_parameters={
            "Ndic":64,              # number of dictionaries
            "Ndim":16,              # input dimension (smaller or equal then Ndic)
            "k":4,                  # number of dictionaries used to create the input
            "tau_m":1,              # in ms
            "dt": 0.1,
            "t_end": 1000,
            "threshold": 0.4,
            "thr_func": None,       # The T(.) function taking two parameters (lambda and u)
            "thr_coeff": 1.,
            "GM":None,
            }

PATH_FIGURES = "./figures/rate_based/"
PATH_CACHING = cmn.HOME+"/data/LCA/tmp/"
cmn.mkdir_p(PATH_FIGURES)
PATH_RESULTS = cmn.HOME+"/data/LCA/rate5/"
cmn.mkdir_p(PATH_RESULTS)
```

```python
seed = 1234567
Ndic = 64
Ndim = 16
GM4 = generative_model_data(Ndic, Ndim, k=4, positive_phi=False, positive_input=False, seed=seed)
GM10 = generative_model_data(Ndic, Ndim, k=10, positive_phi=False, positive_input=False, seed=seed)
```

## Hard threshold - symetric

```python
def hard_threshold_sim(u, thresh):
    vals = u.copy()                                                                   
    vals[(vals<thresh) & (vals>-thresh)]=0.
    return vals

file_name0 = PATH_CACHING + "/rate-based-example5_0.pkl"
```

### k = 4

```python
%%cache $file_name0 a_LCAs0, GM4, pars0, thresholds0

pars0 = default_parameters.copy()
pars0["Ndic"] = Ndic
pars0["Ndim"] = Ndim
pars0["k"] = 4
pars0["GM"] = GM4
pars0["thresh_func"] = hard_threshold_sim

thresholds0 = np.linspace(0,1.5,50)
a_LCAs0 = []
for thr in thresholds0:
    pars0["threshold"] = thr
    a_LCA = simulate(pars0)[:,-1]
    a_LCAs0.append(a_LCA)
```

```python
errors0, nr_active0 = [], []
for a_LCA in a_LCAs0:
    analysis = reconstruction_evaluation(a_LCA, GM4, pars0)
    errors0.append(analysis["opt_error"])
    nr_active0.append(analysis["nonzero"])
```

```python
plot_reconstruction_error(thresholds0, nr_active0, errors0, pars0["Ndic"])
pyl.subplot(1,2,1)
pyl.xlabel(r"threshold ($\lambda$)")
pyl.subplot(1,2,2)
pyl.xlim(.5,1.)
savefig(PATH_FIGURES + "rate-reconstruction-error5_0.png")
```

### k = 4

```python
def soft_threshold_sim(u, thresh):
    vals = u.copy()                                                                   
    vals[(vals<=thresh) & (vals>=-thresh)]=0.
    vals[vals>thresh] -= thresh
    vals[vals<-thresh] += thresh
    return vals

file_name2 = PATH_CACHING + "/rate-based-example5_2.pkl"
```

## Mixed threshold - symetric

```python
def mixed_threshold_sim(u, thresh):
    vals = u.copy()                                                                   
    vals[(vals<=thresh) & (vals>=-thresh)]=0.
    vals[(vals>thresh) & (vals<2*thresh)] -= thresh
    vals[(vals>=0) & (vals<2*thresh)] *= 2.
    vals[(vals<-thresh) & (vals>-2.*thresh)] += thresh
    vals[(vals<=0) & (vals>-2*thresh)] *= 2.
    return vals

file_name4 = PATH_CACHING + "/rate-based-example5_4.pkl"
```
