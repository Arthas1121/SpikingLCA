# A rate based example of the LCA dynamics - generative + hard/soft thresholding

Same as rate-based-example, but this time with the generative model and soft/hard thresholding function comparison:.

### Things I notice:

* Even the hard threshold with the slope 1 will not converge to the optimal solution, if the threshold is too low or if the number of chosen "active" dictionaries is big enough (k > 4).
* Both the soft threshold as well as the hard threshold with the slope under 1 seem to preform worse then the optimal hard threshold.

### When comparing hard Vs. soft threshold with the slope 1 (negative allowed)

* The length of the reconstruction vector in soft thresholding is not optimal. To optimize it, you can not take a simple linear function of the threshold value.

### The threshold/input hypothesis

* I show that changing the threshold is equivalent to multiplying the input with a certain factor, as long as the change factors are the same.

```python
%load_ext ipycache

import neurovivo.plotting.plotting as plt
import matplotlib.pyplot as pyl
import numpy as np
import neurovivo.common as cmn
import brian as br

default_parameters={
            "Ndic":64,              # number of dictionaries
            "Ndim":16,              # input dimension (smaller or equal then Ndic)
            "k":4,                  # number of dictionaries used to create the input
            "v_reset":-60.,         # in mV
            "v_thr":-50.,           # in mV
            "tau_m":1,              # in ms
            "v_rest":-60.,          # in mV
            "dt_delta_thr":10,      # in ms
            "v_rev_e":0.,           # in mV
            "v_rev_i":-80.,         # in mV
            "tau_e":5,              # in ms
            "tau_i":10,             # in ms - note theory requires formally tau_i < tau_e
            "t_refr":5,             # in ms
            "dt": 0.1,
            "t_end": 1000,
            "threshold": 0.4,
            "nonnegative":False,
            "thr_type": "hard",     # "soft" or "hard" threshold.
            "input_factor": 1,
            "thr_coeff": 1.,        
            "seed":None,            # random seed used for the generative model.
            }

PATH_FIGURES = cmn.HOME + "/data/LCA/figures/rate3/"
PATH_FIGURES_REP = "./figures/"
PATH_CACHING = cmn.HOME+"/data/LCA/tmp/"
cmn.mkdir_p(PATH_CACHING)
cmn.mkdir_p(PATH_FIGURES)

def sparsness(vec):
    return 1-1.*np.count_nonzero(vec)/len(vec)

def plot_edge(SMs, i):
    figure()
    plot(SMs[0][i], label="u")
    plot(SMs[1][i], label="a")
    plot(SMs[2][i], label="li")
    legend()

def plot_a_evolutions(SMs, N):
    [plot(SMs[1][i], label=str(i), color = "blue") for i in xrange(N)]
    legend()
    
def plot_u_evolutions(SMs, N):
    [plot(SMs[0][i], label=str(i), color = "red") for i in xrange(N)]
    legend()

def reconstruction_evaluation(a_LCA, GM, pars):

    x_n = GM["x"]/np.linalg.norm(GM["x"])
    rec = np.dot(GM["phi"], a_LCA)
    rec_n = np.zeros(len(rec))
    if not np.linalg.norm(rec) == 0:
        rec_n = rec/np.linalg.norm(rec)
    rec_opt = np.dot(rec_n, x_n) * rec_n

    opt_error = np.linalg.norm(rec_opt - x_n)
    
    result={"normalized error": 1.*np.linalg.norm(GM["x"] - rec)/np.linalg.norm(GM["x"]),
            "normalized a diff": 1.*np.linalg.norm(a_LCA-GM["a"])/np.linalg.norm(GM["a"]),
            "nonzero": np.count_nonzero(a_LCA),
            "error + thr a1": np.linalg.norm(GM["x"] - rec) + pars["threshold"] * np.sum(np.abs(a_LCA)),
            "thr a_t1": pars["threshold"] * np.sum(np.abs(GM["a"])),
            "opt_error": opt_error,
            }
    return result
    
def reconstruction(b_n, SMs, normalize=False):
    rec = np.dot(b_n.T,SMs[1][:][:,-1])
    if np.linalg.norm(rec) != 0 and normalize:
        rec = rec/np.linalg.norm(rec)    
    return rec

def reconstruction_error(b_n, x, SMs, normalize_reconstruction=False):
    rec = reconstruction(b_n, SMs, normalize=normalize_reconstruction)
    return np.linalg.norm(x - rec)

def choose_unique(k, the_list):
    """
    Chose k unique elements of a list the_list. It preserves the order.
    """
    n = len(the_list)
    assert k <= n
    chosen_idx = []
    while len(chosen_idx) < k:
        d = len(chosen_idx)
        new_candidates = list(np.random.random_integers(0, n-1, (k-d)))
        chosen_idx = list(set(chosen_idx + new_candidates))
    chosen_idx.sort()
    return [the_list[i] for i in chosen_idx]

def generative_model_data(Ndic, Ndim, k, seed=None):
    
    if seed:
        np.random.seed(seed=seed)

    b = np.random.randn(Ndic, Ndim)
    norms = np.apply_along_axis(np.linalg.norm, 1, b)
    b_n = b / norms.reshape(-1,1)
    phi = b_n.T
    
    # create a G matrix from the normalized dictionaries
    G = np.zeros((Ndic,Ndic))
    for i in xrange(Ndic):
        for j in xrange(i,Ndic):
            G[i][j] = np.dot(b_n[i], b_n[j])
    G = G + G.T - 2*np.identity(Ndic)
    G[np.abs(G)<0.00000000001] = 0.
    
    # create the sparse vector (ground truth) and use it to generate input vector i_stim
    dict_idx = choose_unique(k, np.arange(Ndic))
    sparse_vector = np.zeros(Ndic)
    for i in dict_idx:
        sparse_vector[i] = np.random.randn()
    
    #sparse_vector_norm = 1.*x/np.linalg.norm(sparse_vector)
    x = np.dot(phi, sparse_vector)
    i_stim = np.dot(phi.T, x)

    result = {"G": G,
              "phi": phi,
              "a": sparse_vector,
              "x": x,
              "i_stim": i_stim,
              }

    return result
```

```python
def simulate_rate(parameters=default_parameters):

    x = cmn.Bunch(parameters)

    GM = generative_model_data(x.Ndic, x.Ndim, x.k, seed=x.seed)
    
    tau_m_inv = 1./(x.tau_m*br.ms)

    br.clear(all=True) # clears away all the brian objects from previous simulations.
    
    def thresh_func(values, threshold, thr_type, thr_coeff, nonnegative):
        vals = values.copy()
        
        if not nonnegative:
            vals[(vals<threshold) & (vals>-threshold)]=0.
            if thr_type == "soft":
                vals[(vals>threshold)] -= threshold
                vals[(vals<-threshold)] += threshold
        else:
            vals[(vals<threshold)]=0.
            if thr_type == "soft":
                vals[(vals>threshold)] -= threshold
            
        return thr_coeff * vals
    
    eqs = '''
          du/dt=(-u-li+i_s)*tau_m_inv     :br.Hz
          a                               :br.Hz
          i_s                             :br.Hz
          li                              :br.Hz
          '''
    
    simulation_clock = br.Clock(dt=x.dt*br.ms)
    recording_clock = br.Clock(dt=1*br.ms)
    
    P = br.NeuronGroup(x.Ndic, model=eqs, threshold=200000*br.Hz, clock=simulation_clock)
    P.i_s = x.input_factor * GM["i_stim"]
    P.a = 0
    
    @br.network_operation(clock=simulation_clock, when="end")
    def set_input(simulation_clock):
        P.li = np.dot(GM["G"], P.a)
        P.a = thresh_func(P.u, x.threshold, x.thr_type, x.thr_coeff, x.nonnegative)
        
    SMs = []
    #SM_1 = br.StateMonitor(P, "u", record=True, timestep=1, clock=recording_clock)
    SM_2 = br.StateMonitor(P, "a", record=True, timestep=1, clock=recording_clock)
    #SM_3 = br.StateMonitor(P, "li", record=True, timestep=1, clock=recording_clock)
    #SM_4 = br.StateMonitor(P, "i_s", record=True, timestep=1, clock=recording_clock)
    #SMs.append(SM_1)
    SMs.append(SM_2)
    #SMs.append(SM_3)
    #SMs.append(SM_4)

    network = br.Network(P, set_input, *SMs)

    # starting the simulation
    network.run(x.t_end*br.ms)

    return 1.*SM_2[:][:,-1]/x.input_factor, GM
```

## Exploring the hypothesis that the change of the threshold is equivalent to changing the input vector size
The way I check this is I choose a standard set of parameters and then simulatneously vary the threhsold and the factor with which I multiply the selected input in a way that I keep their ratio the same. If this doesn't change the results, we have the hypothesis confirmed.

```python
%%cache /Users/mpelko/data/LCA/tmp/rate-based-example3_0_hard.pkl a_LCAs0_h, GMs0_h, pars0_h, thresholds0_h
pass
```

```python
pars2 = pars0_h.copy()

input_factors = [0.1, 0.2, 0.4, 0.6, 0.8, 1., 1.25, 1.5, 2., 4., 8., 16.]
a_LCAs2, GMs2 = [], []
for i_f in input_factors:
    pars2["threshold"] = 0.2*i_f
    pars2["input_factor"] = i_f
    a_LCA, GM = simulate_rate(pars2)
    a_LCAs2.append(a_LCA)
    GMs2.append(GM)
```

```python
analysis2 = []
for a_LCA, GM in zip(a_LCAs2, GMs2):
    analysis2.append(reconstruction_evaluation(a_LCA, GM, pars2))
```

```python
figure(figsize=(16,5))
subplot(1,3,1)
plot(input_factors, [an["normalized error"] for an in analysis2])
xlim(0,1)
xlabel(r"$\lambda$")
xlabel(r"$\lambda$")
subplot(1,3,2)
plot(input_factors, [1.-1.*an["nonzero"]/pars2["Ndic"] for an in analysis2])
plot(input_factors, [1.-1.*pars2["k"]/pars2["Ndic"]]*len(input_factors), "gray", linestyle="--")
xlim(0,1)
xlabel(r"$\lambda$")
subplot(1,3,3)
plot(input_factors, [an["normalized a diff"] for an in analysis2])
xlabel(r"$\lambda$")
```

We have the hypothesis confirmed. Changing the threshold is effectively same as multiplying the input with some factor.
